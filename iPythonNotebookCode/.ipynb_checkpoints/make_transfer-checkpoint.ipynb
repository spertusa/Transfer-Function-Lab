Reads in the two files, "bandlimited.txt" and "data11.txt" into arrays s and r. These are mockups of the input and output from the circuit under test. We want to turn them into harmonic spectra, divide them and plot the absolute value of the ratio.

Here is the start of this that reads the two files in from the local directory.

```python
%pylab inline
import numpy.fft as fft
import glob as gl
import os
import json
from pprint import pprint
```

```python
directory='/Users/Ayelet/Desktop/NewArduino/ProcessingFiles/StructuredReceive3_bin'
datafiles=gl.glob(os.path.join(directory,'Data*.txt'))
#datafiles=datafiles[0:-1]
datafiles
```

```python
s=genfromtxt("bandlimited.txt") # get the source array and average every other point
r=np.zeros(1024)
num = 0
for data_file in datafiles:
    ra=genfromtxt(data_file) #get the result array
    if (ra.size == 1024):
        r += ra
        num += 1
r/=num
ra -= r
```

```python
plot(s)
plot(1000*ra)
plot(r)
pylab.xlabel("Frequency(Hz)")
pylab.ylabel("Transfer Funtion")
```

```python
ss=fft.rfft(s) #fourier transform the source array
rr=fft.rfft(r) #fourier transform the result array
t=rr/ss #divide them by each other
df=2450./(2*1024.) #calculate the frequency steps like we discussed today
f=np.arange(0,512)*df #make the frequency array
c = 0.99e-7 #value of capicator in Farads
res = 1.2e4  #value of resistor in Ohms
m = np.abs(t[10])/(1+1j*res*c*(2*np.pi*f)) #model of ideal transfer funtion in green
loglog(f,np.abs(t[0:512])) #plot it up in a log/log plot.
loglog(f,np.abs(m))
pylab.xlabel("Frequency(Hz)")
pylab.ylabel("Transfer Funtion")
```
